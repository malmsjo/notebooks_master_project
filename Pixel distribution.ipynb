```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import glob
import os
from os import *
import re
import imutils
from skimage.feature import peak_local_max
from skimage.morphology import watershed
from scipy import ndimage
import collections
import random
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text
def natural_keys(text):
    return [atoi(c) for c in re.split('(\d+)', text)]
```

```python
image_list = []
    

for filename in sorted(glob.glob('/mnt/holomonitor-pvc/project/A549Costar/Well F3*.tiff'), key=natural_keys): 
    im=cv2.imread(filename)
    imarray = np.array(im)
    #gray_image = cv2.cvtColor(imarray, cv2.COLOR_BGR2GRAY)
    image_list.append(imarray)
```
