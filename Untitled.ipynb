```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import glob
import os
from os import *
import re
import imutils
from skimage.feature import peak_local_max
from skimage.morphology import watershed
from scipy import ndimage
import collections
import random
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text
def natural_keys(text):
    return [atoi(c) for c in re.split('(\d+)', text)]
```

```python
p = '/home/jovyan/send_ola_1/'

t = []
for i in sorted(listdir(p),key=natural_keys ):
    im =cv2.imread(p+i)
    if np.count_nonzero(im == 255) < 1000:
        t.append(im)
```

```python
img = cv2.imread('/home/jovyan/A549-TL-pilot_C03_s15DA8D61F-A330-4B36-B171-1DB8D1E88898.tif')
plt.imshow(img)
```

```python
shifted = cv2.pyrMeanShiftFiltering(img, 10, 50)
plt.imshow(shifted)
```

```python
gray = cv2.cvtColor(shifted,cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=20, labels=thresh)

markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
plt.imshow(labels)
```

```python
ar = []
cx = []
cy = []
for label in np.unique(labels):
    if label == 0:
        continue
        
    
    mask = np.zeros(gray.shape, dtype='uint8')
    mask[labels == label] = 255
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    
    
    cnts = imutils.grab_contours(cnts)
    c = max(cnts, key=cv2.contourArea)
    area = cv2.contourArea(c)
    ((x, y), r) = cv2.minEnclosingCircle(c)
    cx.append(int(x))
    cy.append(int(y))
    ar.append(area)
    if area > 1000:

        cv2.drawContours(img, [c], -1,(0, 255, 0), 2)
    else:
        cv2.drawContours(img, [c], -1,(255,0,0), 2)
        #cv2.circle(img, (int(x), int(y)), int(r), (0, 255, 0), 2)
    
plt.imshow(img)
```
